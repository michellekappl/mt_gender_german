```python
fst = []
female_jobs_idx = {}
male_jobs_idx = {}
female_jobs_idx_2cnd = {}
male_jobs_idx_2cnd = {}

female_jobs_2cnd = []
female_jobs = []
male_jobs_2cnd = []
male_jobs = []
scnd = []

for idx, line in enumerate(open('./datasets/de_min_new.txt', encoding="utf8")):
    sen = line.strip().split('\t')
    job = sen[3]
    gender = sen[0]
    if idx > 151:
        if job not in scnd:
            scnd.append(job)
        if gender == "female" and job not in female_jobs_2cnd:
            female_jobs_idx_2cnd[job] = [idx]
            female_jobs_2cnd.append(job)
        elif gender == "female":
            female_jobs_idx_2cnd[job].append(idx)
        if gender == "male" and job not in male_jobs_2cnd:
            male_jobs_idx_2cnd[job] = [idx]
            male_jobs_2cnd.append(job)
        elif gender == "male":
            male_jobs_idx_2cnd[job].append(idx)
    else:
        if gender == "female" and job not in female_jobs:
            female_jobs_idx[job] = [idx]
            female_jobs.append(job)
        elif gender == "female":
            female_jobs_idx[job].append(idx)
        if gender == "male" and job not in male_jobs:
            male_jobs_idx[job] = [idx]
            male_jobs.append(job)
        elif gender == "male":
            male_jobs_idx[job].append(idx)
        if job not in fst:
            fst.append(job)

all_jobs = fst + list(set(scnd) - set(fst))

import pandas as pd

job_data = {
    'job': [],
    'gender': [],
    '2cnd': [], 
    'idx': []
}

for idx, job in enumerate(all_jobs):
    if job in fst:
        if job in female_jobs:
            job_data['job'].append(job)
            job_data['idx'].append(female_jobs_idx[job])
            job_data['gender'].append('female')
            job_data['2cnd'].append('no')
        if job in male_jobs:
            job_data['job'].append(job)
            job_data['idx'].append(male_jobs_idx[job])
            job_data['gender'].append('male')
            job_data['2cnd'].append('no')
    if job in scnd:
        if job in female_jobs_2cnd:
            job_data['job'].append(job)
            job_data['idx'].append(female_jobs_idx_2cnd[job])
            job_data['gender'].append('female')
            job_data['2cnd'].append('yes')
        if job in male_jobs_2cnd:
            job_data['job'].append(job)
            job_data['idx'].append(male_jobs_idx_2cnd[job])
            job_data['gender'].append('male')
            job_data['2cnd'].append('yes')

print(job_data)
df = pd.DataFrame(job_data)

df.to_csv('job_data_temp.csv', sep=';', index=False)
print(df)
```

```python
aba = pd.read_csv('./statistics_dataset/datenbank_full.csv', sep=';')
aba.columns = ['gender', 'ab_code', 'name', 'n1', 'number', 'bla']
aba = aba.drop(['n1', 'bla'], axis=1)
jobs = pd.read_csv('job_data_temp.csv', sep=';')
jobs.columns = ['job', 'gender' ,'2cnd', 'idx', 'ab_code']

filtered = aba[aba['ab_code'].isin(jobs['ab_code'])]

men = filtered[filtered['gender'].str.contains('Männer')]
women = filtered[filtered['gender'].str.contains('Frauen')]

import re
import numpy as np
    
def toInt(str):
    return int(''.join(re.findall(r'\d+', str)))

concat = men.drop(['gender'], axis=1)
concat = concat.rename(columns = {'number':'men'})
concat['women'] = women['number'].values

concat['men'] = concat['men'].apply(toInt)
concat['women'] = concat['women'].apply(toInt)

concat['total'] = concat['men'] + concat['women']
concat['perc_women'] = concat['women'] / concat['total'] * 100
concat['perc_men'] = concat['men'] / concat['total'] * 100

concat['prominent_gender'] = np.where(concat['men'] > concat['women'], 'male', 'female')
```

```python
import pandas as pd
import docx2txt

de = pd.read_csv('./datasets/de_min_new.txt', sep='\t', names=['gender', 's_index', 'sentence', 'subject'])

with open('../../data/aggregates/pure_txt.txt', 'w') as txt:
    for sentence in de['sentence']:
        txt.write(sentence + '\n')

def translations_convert(mt, transl_f):
    text = docx2txt.process('../../translations/' + mt + '/translations/' + transl_f + '.docx')
    text = text.split('\n')
    text = [line for line in text if line != '']

    for idx, line in enumerate(text):
        print(idx, line)
        line = de.iloc[idx]['sentence'] + ' ||| ' + line + '\n'
        text[idx] = line
    with open('../../translations/' + mt + '/de-' + transl_f + '.txt', 'w') as f:
        f.writelines(text)

#for lang in ['it', 'fr', 'uk', 'ru', 'es', 'he', 'ar']:
#    translations_convert(lang)

for model in ['bing', 'google', 'systran']:
    for lang in ['it', 'fr', 'uk', 'ru', 'es', 'he', 'ar']:
        translations_convert(model, lang)
```
